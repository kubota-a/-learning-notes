## 問題1. 

数字の7と3に対して、以下の出力になるように四則演算をおこなってください。

割り算については、①余りを出力するパターンと②割り切るパターンの2通り計算しましょう。

```
7 + 3 = 10
7 - 3 = 4
7 × 3 = 21
7 ÷ 3 = 2 余り 1
7 ÷ 3 = 2.3333333333333335
```

````python
a = 7
b = 3
print(f'{a} + {b} = {a + b}')    
print(f'{a} - {b} = {a - b}')
print(f'{a} × {b} = {a * b}')
print(f'{a} ÷ {b} = {a // b} 余り {a % b}')
print(f'{a} ÷ {b} = {a / b}')
````

## 問題2. 

好きな文字を入力して、以下のように入力した文字を出力するプログラムを作成しましょう。

```
好きな文字を入力してください > Python
入力された文字 : Python
```


````python
strig = input("好きな文字を入力してください > ")    # 好きな文字を入力してください > ▢　・この▢にPythonと入力してEnterすると、
print(f'入力された文字 :  {strig}')    # 入力された文字 : Python　と出力される
````
